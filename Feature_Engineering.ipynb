```python
import pandas as pd
import numpy as np
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
pd.pandas.set_option('display.max_columns',None)
```

```python
df = pd.read_csv("train.csv")
```

```python
df.head()
```

```python
#also preform the same forthe test data
#lets compare the missing values

features_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtypes == 'O']

for feature in features_nan:
    print("{} has {}% missing values".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
# replace missing values
#here we have created a new label Missing
def replace_cat_feature(df,feature_nan):
    data = df.copy()
    data[feature_nan] = data[feature_nan].fillna('Missing')
    return data

df = replace_cat_feature(df,features_nan)
df[features_nan].isnull().sum()
```

```python
df.head()
```

```python
#now lets check for numerical values that contain missing values
numerical_with_nan = [feature for feature in df.columns if df[feature].isnull().sum()>1 and df[feature].dtype != 'O']

#we will print the numerical nan values annd percentage of missing values
for feature in numerical_with_nan:
    print("{}: {}% missing values ".format(feature,np.round(df[feature].isnull().mean(),4)))
```

```python
#replacing the missing numerical vull values
for feature in numerical_with_nan:
    median_value = df[feature].median()
    
    #create a new feature to capture nan values
    df[feature+'nan'] = np.where(df[feature].isnull(),1,0)
    df[feature].fillna(median_value,inplace=True)
    
df[numerical_with_nan].isnull().sum()
```

```python
df.head(25)
```

```python
# now we deal wiith the temporal vars
#here we didnt consider the yrsold beacuase the orice was reducing for that feature
for feature in ['YearBuilt','YearRemodAdd','GarageYrBlt']:
    df[feature] = df["YrSold"]- df[feature]
```

```python
df.head()
```

```python
df[['YearBuilt','YearRemodAdd','GarageYrBlt']].head()
```

# Numerical Values
## Since newmerical values are skewed (has many outliers), we will use log normal dist

```python
num_features = ["LotFrontage","LotArea","1stFlrSF","GrLivArea","SalePrice"]

for feature in num_features:
    df[feature] = np.log(df[feature])
```

```python
df.head()
```

## Handling rare cat features
### We will remove the obs which are present in less than 1% of the obs in their category

```python
cat_feature = [feature for feature in df.columns if df[feature].dtype == "O"]
cat_feature
```

```python
#we can skip obs with less than 1% cause they wont affect with that much weight, we will just convert them into category
for feature in cat_feature:
    temp = df.groupby(feature)["SalePrice"].count()/len(df)
    temp_df = temp[temp>0.01].index
    df[feature] = np.where(df[feature].isin(temp_df),df[feature],"Rare_var")
```

```python
df.head(25)
```

```python
for feature in cat_feature:
    labels_ordered = df.groupby([feature])['SalePrice'].mean().sort_values().index
    labels_ordered={k:i for i,k in enumerate(labels_ordered,0)}
    df[feature]=df[feature].map(labels_ordered)
```

```python
df.head(10)
```

## Feature Scaling

```python
feature_scale = [feature for feature in df.columns if feature not in ["Id","SalePrice"]]
```

```python
len(feature_scale)
```

```python
from sklearn.preprocessing import MinMaxScaler
scaler = MinMaxScaler()
```

```python
scaler.fit(df[feature_scale])
```

```python
# we didnt use standard scaler as there were not many negative values
# min max scaler scales values between 0 and1 
# standard scaler converts values in std normal distribution
data = pd.concat([df[["Id","SalePrice"]].reset_index(drop=True),
                     pd.DataFrame(scaler.transform(df[feature_scale]),columns=feature_scale)],axis=1)
```

```python
data.head()
```

```python
data.to_csv("X_train.csv",index=False)
```
