```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
from sklearn.linear_model import Lasso
from sklearn.feature_selection import SelectFromModel

#to viz the columns in the data frame
pd.pandas.set_option("display.max_columns",None)
```

```python
df = pd.read_csv("x_train.csv")
```

```python
df.head()
```

```python
y_train = df["SalePrice"]
```

```python
x_train = df.drop(["Id","SalePrice"],axis=1)
```

# Applying Feature Selection

```python
# firtly we specify the lasso selection model and select a suitable alpha (equivalent of penalty)
# we know that bigger the alpha less the umber of features will be selected
# then we use selectfrommodel from sklearn which selects the features whose coeff is non zero


# lasso penalizes features which have higher weights
#select from model, selects features based on the importanc eof weights
feature_sel_model = SelectFromModel(Lasso(alpha=0.005,random_state=0)) #remember to set the seed, 
feature_sel_model.fit(x_train,y_train)
```

```python
feature_sel_model.get_support()
#here true basically means that this feature is important and false indicates otherwise
```

```python
#lets print the total and number of selected features
sel_features = x_train.columns[(feature_sel_model.get_support())]

print("Total Features : {}".format(x_train.shape[1]))
print("Selected Features: {}".format(len(sel_features)))
print("Features with coeff shrank to 0: {}".format(np.sum(feature_sel_model.estimator_.coef_ == 0)))
```

```python
sel_features
```

```python
x_train = x_train[sel_features]
```

```python
x_train
```

```python
from sklearn.ensemble import RandomForestRegressor
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
from sklearn.model_selection import train_test_split
import joblib
```

```python
x_train.shape
```

```python
y_train.shape
```

```python
X_train, X_val, Y_train, Y_val = train_test_split(x_train, y_train, test_size=0.2, random_state=42)
```

```python
model = RandomForestRegressor(n_estimators=100, random_state=42)
model.fit(X_train, Y_train)
```

```python
y_pred_train = model.predict(X_train)
y_pred_val = model.predict(X_val)
```

```python
# Training set evaluation
mae_train = mean_absolute_error(Y_train, y_pred_train)
mse_train = mean_squared_error(Y_train, y_pred_train)
r2_train = r2_score(Y_train, y_pred_train)

# Validation set evaluation
mae_val = mean_absolute_error(Y_val, y_pred_val)
mse_val = mean_squared_error(Y_val, y_pred_val)
r2_val = r2_score(Y_val, y_pred_val)

print("Training set evaluation:")
print(f"MAE: {mae_train}")
print(f"MSE: {mse_train}")
print(f"R-squared: {r2_train}")

print("\nValidation set evaluation:")
print(f"MAE: {mae_val}")
print(f"MSE: {mse_val}")
print(f"R-squared: {r2_val}")
```

```python
joblib.dump(model, 'random_forest_model.pkl')
```

```python
from sklearn.model_selection import train_test_split, GridSearchCV
```

```python
param_grid = {
    'n_estimators': [100, 200, 300],
    'max_features': ['auto', 'sqrt', 'log2'],
    'max_depth': [10, 20, 30, None],
    'min_samples_split': [2, 5, 10],
    'min_samples_leaf': [1, 2, 4]
}

grid_search = GridSearchCV(estimator=RandomForestRegressor(random_state=42),
                           param_grid=param_grid,
                           cv=3,
                           n_jobs=-1,
                           verbose=2)

grid_search.fit(X_train, Y_train)
```

```python
best_model = grid_search.best_estimator_
best_model.fit(X_train, Y_train)
```

```python
y_pred_train = best_model.predict(X_train)
y_pred_val = best_model.predict(X_val)
```

```python
X_train.shape
```

```python
# Training set evaluation
mae_train = mean_absolute_error(Y_train, y_pred_train)
mse_train = mean_squared_error(Y_train, y_pred_train)
r2_train = r2_score(Y_train, y_pred_train)

# Validation set evaluation
mae_val = mean_absolute_error(Y_val, y_pred_val)
mse_val = mean_squared_error(Y_val, y_pred_val)
r2_val = r2_score(Y_val, y_pred_val)

print("Training set evaluation:")
print(f"MAE: {mae_train}")
print(f"MSE: {mse_train}")
print(f"R-squared: {r2_train}")

print("\nValidation set evaluation:")
print(f"MAE: {mae_val}")
print(f"MSE: {mse_val}")
print(f"R-squared: {r2_val}")
```

```python
joblib.dump(best_model, 'random_forest_model_best.pkl')
```
